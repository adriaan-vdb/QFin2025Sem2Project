```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv("prices.csv")
```

```python
df.head()
```

```python
df['Mids'] = (df["Bids"] + df["Asks"])/2
```

```python
plt.plot(df['Mids'])
```

```python
a = np.exp(np.random.normal(0, .001, 1000))
plt.plot(a)
plt.show()




```

```python
## Calculating returns
ticker = 5
df["Trailing_Returns_5"] = (df["Mids"] - df ["Mids"].shift(ticker)) / df ["Mids"].shift(ticker)

df["Forward_Returns_5"] = (df["Mids"].shift(-ticker) - df["Mids"]) / df["Mids"]
```

```python
# Remove NaN values for plotting
df_clean = df.dropna()

plt.plot(df_clean["Trailing_Returns_5"], df_clean["Forward_Returns_5"], "o", alpha=0.6)
plt.xlabel("Trailing Returns (5-period)")
plt.ylabel("Forward Returns (5-period)")
plt.title("Trailing vs Forward Returns")
plt.grid(True, alpha=0.3)
plt.show()
```
